```python
# New feature i.e %growth from 2008 to 2009
miss_bool = data['C1) Funding FY2008']!= 0
data.loc[miss_bool,'growthper'] = ((data.loc[miss_bool,'C2) Funding FY2009'] - data.loc[miss_bool,'C1) Funding FY2008'])/data.loc[miss_bool,'C1) Funding FY2008'])*100
```

```python
# %growth for cases having zero funding in 2008
miss_bool2 = (data['C1) Funding FY2008'] == 0) & (((data['C2) Funding FY2009']) - (data['C1) Funding FY2008'])) > 0)
data.loc[miss_bool2,'growthper'] = 1
miss_bool3 = (data['C1) Funding FY2008'] == 0) & (((data['C2) Funding FY2009']) - (data['C1) Funding FY2008'])) == 0)
data.loc[miss_bool3,'growthper'] = 0
```

```python
# creating target variable
bool = data['growthper']>0
data.loc[bool,'target'] = 1
bool2 = data['growthper']<=0
data.loc[bool2,'target'] = 0
```

* For all the programs for which Funding FY2008 and Funding FY2009 are not Known I have consider them not attending the quest     and alloted 0 to them.
* For all the programs for which Funding FY2008 are not known but Funding FY2008 are known I have alloted 1 to them.

```python
def hypothesis(x,y):
    boolean = data[x].isnull() & data[y].isnull()
    data.loc[boolean,'target'] = 0
    boolean2 = data[x].isnull() & data[y].notnull()
    data.loc[boolean2,'target'] = 1
```
